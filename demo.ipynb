```python
# Extract all unique tags with their tag IDs
all_tags = []
for tag_list in df['tag']:
    if isinstance(tag_list, list):
        for tag_dict in tag_list:
            all_tags.append((tag_dict['tagid'], tag_dict['tagname']))

# Get unique tags
unique_tags = list(set(all_tags))
unique_tags_sorted = sorted(unique_tags, key=lambda x: x[0])

# Create a dataframe for better display
tags_df = pd.DataFrame(unique_tags_sorted, columns=['tag_id', 'tag_name'])
tags_df
```

```python
# Extract all unique options with their option IDs
all_options = []
for option_list in df['option']:
    if isinstance(option_list, list):
        for option_dict in option_list:
            all_options.append((option_dict['answerid'], option_dict['optiontext']))

# Get unique options
unique_options = list(set(all_options))

unique_options_sorted = sorted(unique_options, key=lambda x: x[0])


# Create a dataframe for better display
options_df = pd.DataFrame(unique_options_sorted, columns=['option_id', 'option_text'])
options_df
```

```python
#Extract all questions in a list
all_questions = []
for ques_dict in df['question']:
    if isinstance(ques_dict,dict):
       all_questions.append((ques_dict['questionid'] , ques_dict['questiontext'] , ques_dict['questiontype']))

#Get all unique questions
unique_questions = set(all_questions)

#Sort all questions by its questionid
unique_questions_sorted = sorted(unique_questions , key = lambda x:x[0])

# Create a dataframe for better display
questions_df = pd.DataFrame(unique_questions_sorted, columns=['question_id', 'question_text', 'question_type'])
questions_df
```

```python
# Define a function to get correct answer for each row
def get_correct_ans(option_list):
    if isinstance(option_list, list):
        for option_dict in option_list:
            if option_dict.get('answerstatus') == 'Correct':
                return option_dict.get('optiontext')
    return None

# Apply the function to create a new column
df['Correct_Ans'] = df['option'].apply(get_correct_ans)

# Display the result
df[['question', 'Correct_Ans']].head(10)
```
